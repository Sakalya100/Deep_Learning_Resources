```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
# Independent Features
X = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
X.head()
```

```python
# Dependent feature
y = iris.target
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

In Decision tree we dont need any normalization or Standadrization. It can handle it 

```python
from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train, y_train)
```

```python
## Visualise the decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier, filled=True)
```

```python
y_pred = treeclassifier.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```

```python
## Post Pruning
# Set depth upto 2 levels only
treeclassifier = DecisionTreeClassifier(max_depth=2)
```

```python
treeclassifier.fit(X_train, y_train)
```

```python
tree.plot_tree(treeclassifier, filled=True)
```

```python
y_pred = treeclassifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```

## So we can conclude that if the max_depth is till the last level then the accuracy is better but it also may lead to overfitting

```python
params
```
